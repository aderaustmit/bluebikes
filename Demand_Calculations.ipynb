```python
#import detailed stations csv file

stations_long_df = pd.read_csv('Bluebikes.csv', header=0)

#filter to Salem
#stations_df[stations_df.District == "Salem"]
stations_long_df

#import stations csv file with district info

stations_df = pd.read_csv("Station Data.csv", header = 1)
stations_df
```

```python
#merge so we can get lookup key
stations_key = pd.merge(stations_long_df, stations_df,
                                left_on="short_name",
                                right_on="Number",
                                how="inner")

salem_key = stations_key[stations_key.District == "Salem"]
salem_key

salem_ids = salem_key[["station_id","Number","Name","Latitude","Longitude","Total docks"]]
salem_ids
```

```python
#demand from trip histories data

all_months = pd.DataFrame()
months_df = [trips_sep22, trips_oct22, trips_aug22, trips_jul22, trips_jun22, trips_may22, trips_apr22, trips_mar22, trips_feb22, trips_jan22, trips_dec21, trips_nov21]

for trips_df in months_df:
    start_stations = pd.DataFrame().assign(station_id = trips_df['start_station_id'])
    end_stations = pd.DataFrame().assign(station_id = trips_df['end_station_id'])
    start_end_stations = pd.concat([start_stations,end_stations]) #combine start and end stations into one column
    
    #get counts of station uses
    station_counts = start_end_stations.groupby(['station_id'])['station_id'].count().reset_index(name="count")
    
    #append to total list of stations from all months
    all_months = pd.concat([all_months, station_counts])

all_months
```

```python
#get counts of station uses across all months
station_counts_all_months = all_months.groupby(['station_id'])['count'].sum().reset_index(name="count")
station_counts_all_months
```

```python
#merge station capacity data with station count data
salem_count_cap = pd.merge(salem_ids, station_counts_all_months,
                               left_on="station_id",
                               right_on="station_id",
                               how="left")

salem_demand = salem_count_cap.assign(demand = salem_count_cap['count']/salem_count_cap['Total docks'])
salem_demand

#write to csv
#salem_demand.to_csv('salem_demand_allmonths_df')
```
